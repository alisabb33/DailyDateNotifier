```python
# Generate scaled and exportable CAD-style + printable floor plan for a 30' x 28' garage smokehouse.

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrow

# Dimensions in feet

length = 30
width = 28

fig, ax = plt.subplots(figsize=(10, 9))

# Outer walls

ax.add_patch(Rectangle((0, 0), width, length, fill=None, edgecolor='black', linewidth=2))

# Zone divisions (y = distance from front door)

packaging_y = 8
smoke_y = 20
cold_y = 28

# Draw internal dividers

ax.plot([0, width], [packaging_y, packaging_y], 'gray', linestyle='--', linewidth=1)
ax.plot([0, width], [smoke_y, smoke_y], 'gray', linestyle='--', linewidth=1)

# Zone labels

ax.text(width/2, 4, 'PACKAGING & FINISHED GOODS ZONE (8 ft)', ha='center', va='center', fontsize=10)
ax.text(width/2, 14, 'SMOKING & PREP ZONE (12 ft)', ha='center', va='center', fontsize=10)
ax.text(width/2, 24, 'COLD STORAGE & CLEAN-UP ZONE (8 ft)', ha='center', va='center', fontsize=10)

# Equipment (approx positions)

ax.text(4, 24, '3-Comp Sink', fontsize=9, color='blue')
ax.text(10, 24, 'Handwash Sink', fontsize=9, color='blue')
ax.text(20, 24, 'Fridge / Freezer', fontsize=9, color='blue')
ax.text(5, 14, 'Smoker / Dehydrator', fontsize=9, color='red')
ax.text(15, 14, 'Prep Tables', fontsize=9, color='red')
ax.text(10, 4, 'Vacuum Sealer / Labeling', fontsize=9, color='green')
ax.text(18, 4, 'Storage Shelving', fontsize=9, color='green')

# Flow arrows

ax.add_patch(FancyArrow(width/2, 26, 0, -6, width=0.4, head_width=1.5, head_length=1.5, fc='black', ec='black'))
ax.add_patch(FancyArrow(width/2, 18, 0, -6, width=0.4, head_width=1.5, head_length=1.5, fc='black', ec='black'))
ax.text(width/2 + 2, 15, 'Flow: Raw → Smoked → Packaged → Stored', fontsize=8, rotation=90, va='center')

# Plot formatting

ax.set_xlim(0, width)
ax.set_ylim(0, length)
ax.set_aspect('equal')
ax.set_xlabel('Width (ft)')
ax.set_ylabel('Length (ft)')
ax.set_title('Micro Smokehouse Floor Plan (30 ft x 28 ft)')
ax.axis('off')

plt.tight_layout()

# Save export versions

plt.savefig('smokehouse_layout.pdf')  # Printable PDF
plt.savefig('smokehouse_layout.svg')  # Vector CAD-style export
plt.savefig('smokehouse_layout.png', dpi=300)  # High-res image

plt.show()

print('Exported files: smokehouse_layout.pdf, smokehouse_layout.svg, smokehouse_layout.png')
```
